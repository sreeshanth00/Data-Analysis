```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print('Setup Complete')
```

```python
df = my_data 
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
df.sort_values(by='2022 Population', ascending =False).tail()
```

```python
df.sort_values(by='World Population Percentage', ascending =False).head()
```

```python
df.corr(numeric_only =True)
```

```python
sns.heatmap(df.corr(numeric_only =True),annot =True)
plt.figure(figsize=(20,7))
plt.show()
```

```python
df.groupby('Continent').mean(numeric_only =True).sort_values(by='Rank')
```

```python
dfm=df.groupby('Continent')[['2022 Population', '2020 Population', '2015 Population',
       '2010 Population', '2000 Population', '1990 Population',
       '1980 Population', '1970 Population']].sum()
dfm
```

World percentage by Continents

```python
dfpie = df.groupby('Continent')[['World Population Percentage']].sum().sort_values(by='World Population Percentage',ascending =False)
dfpie
```

```python
plt.figure(figsize=(16,9))
plt.pie(dfpie['World Population Percentage'], labels = dfpie.index.tolist(),autopct='%1.1f%%')
plt.show()
```

## Top 10 Countries by Population

```python
plt.figure(figsize=(16,9))
sns.barplot(df.sort_values(by='2022 Population',
            ascending = False).head(10),x='Country',
            y='2022 Population',hue='Continent').set(
            title='Top 10 Countries by Population',
            ylabel="in Billions" ,
            xlabel="by Country")
plt.show()
```

## 10 Least populas Countries

```python
plt.figure(figsize=(16,9))
plt.xticks(rotation=25)
sns.barplot(df.sort_values(by='2022 Population',
            ascending = True).head(10),x='Country',
            y='2022 Population',hue='Continent').set(
            title='Top 10 Countries by Population',
            ylabel="in Billions" ,
            xlabel="by Country",
            )
plt.show()
```
